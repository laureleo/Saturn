```python
split = text.split(sep = '.')
newsplit = [s.strip() for s in split]
nonulls = []
for s in newsplit:
    if len(s) == 0:
        continue
    else:
        nonulls.append(s)
        
noQuotes = []
for s in nonulls:
    s2 = s.replace('“', "")
    s3 = s2.replace('”', "")
    noQuotes.append(s3.strip())
    
english_sentences = [translator.translate(sentence).text for sentence in noQuotes]

literals = ['"' + s + '"' for s in english_sentences]
```

```python
import random as rand
literals = literals[84:]
```

```python
def analyze_text():
    from selenium import webdriver
    match_url = []
    match_sentence = []

    # prepare the option for the chrome driver
    options = webdriver.ChromeOptions()
    #options.add_argument('headless')

    # start chrome browser
    browser = webdriver.Chrome(chrome_options=options)
    for literal in literals:
        time.sleep(rand.uniform(0.1, 5))
        print(f"checking {literal} ")
        browser.get(f"https://google.com/search?q={literal}")
        #input()
        #print(browser.current_url)
        source = browser.page_source
        soup = BeautifulSoup(source)
        if (soup.find('form')['id'] =='captcha-form'):
            print("Whoops, the CAPTCHA police caught you! Quit, and try again later, or make the code better")
            df = pd.DataFrame()
            df['Sentence'] = match_sentence
            df['link'] = match_url
            browser.quit()

            return df

        subsoup = soup.find('div', attrs = {'id':'topstuff'})
        result = subsoup.text.split()[:3]
        if (result == ['Inga', 'resultat', 'hittades']):
            print("No match")
        else:
            print("Match found!")
            match_url.append(browser.current_url)
            match_sentence.append(literal)

    browser.quit()
    

    df = pd.DataFrame()
    df['Sentence'] = match_sentence
    df['link'] = match_url
    return df
```

```python
df = analyze_text()
```
