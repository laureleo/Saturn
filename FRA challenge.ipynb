```python
import pandas as pd
import json as js
```

```python
data_path ='assets/fra/data1.json'
try:
    pd.read_json(data_path)
except:
    jsonlist = {}
    file = open(data_path, 'r') 
    lines = file.readlines() 

    count = 0
    # Strips the newline character 
    for line in lines: 
        jsonlist[count] = js.loads(line.strip())
        count+=1
```

```python
df = pd.DataFrame.from_dict(jsonlist, orient='index')
df.head(2)
```

```python
def adressSplitter(ip):
    parts = ip.split('.')
    return parts
    
#adressSplitter('3.3.3.2')
```

```python
subssetdf = df.head(100)

def dfAdressSplitter(df, prefix):
    df[prefix + '1'] = 0
    df[prefix + '2'] = 0
    df[prefix + '3'] = 0
    df[prefix + '4'] = 0
    
    
    #df['1'] = df.apply(lambda x: adressSplitter(x[prefix]), axis = 1)
    df[prefix + '1'] = df.apply(lambda x: adressSplitter(x[prefix])[0], axis = 1)
    df[prefix + '2'] = df.apply(lambda x: adressSplitter(x[prefix])[1], axis = 1)
    df[prefix + '3'] = df.apply(lambda x: adressSplitter(x[prefix])[2], axis = 1)
    df[prefix + '4'] = df.apply(lambda x: adressSplitter(x[prefix])[3], axis = 1)
    return df

expandedSRC = dfAdressSplitter(df, 'ip_src')
expandedDST = dfAdressSplitter(expandedSRC, 'ip_dst')
```
