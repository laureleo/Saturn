```python
from bs4 import BeautifulSoup
import urllib.request
parser = 'html.parser'  # or 'lxml' (preferred) or 'html5lib', if installed
```

```python
# Get the most popular artists
urls = []
popularartistsurl = f"http://www.acclaimedmusic.net/061024/1948-09art.htm"
urls.append(popularartistsurl)
for i in range(2,6):
    popularartistsurl = f"http://www.acclaimedmusic.net/061024/1948-09art{i}.htm"
    urls.append(popularartistsurl)

```

```python
popular_artists = []
for url in urls:
    resp = urllib.request.urlopen(url)
    soup = BeautifulSoup(resp, parser, from_encoding=resp.info().get_param('charset'))
    for link in soup.find_all('a', href=True):
        if not link.text[0].isnumeric():
                if not link.text[0] == '(':
                    if not link.text == "Albums":
                        if not link.text == "Songs":
                            popular_artists.append(link.text)
```

```python
mbzurls = []
for artist in popular_artists:
    urlcompliantname = urllib.parse.quote(artist)
    mbzurl = f"https://musicbrainz.org/search?query={urlcompliantname}&type=artist&method=indexed"
    mbzurls.append(mbzurl)
```

```python
popularIDS = []
for mbzurl in mbzurls:
    resp = urllib.request.urlopen(mbzurl)
    soup = BeautifulSoup(resp, parser, from_encoding=resp.info().get_param('charset'))
    
    firstTaken = False
    
    for link in soup.find_all('a', href=True):

        if link['href'][:8] == "/artist/":
            popularIDS.append(link['href'][8:])
            break;

```

```python
with open('top1000Artists.txt', 'w') as f:
    for item in popularIDS:
        f.write("%s\n" % item)
```
